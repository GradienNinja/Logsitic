```python

import kagglehub

# Download latest version
path = kagglehub.dataset_download("taweilo/loan-approval-classification-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/loan-approval-classification-data/loan_data.csv")
```

```python
data.head()
```

```python
data["person_gender"] = data["person_gender"].map({"female" : 0,"male" : 1})
```

```python
data.head()
```

```python
categorical_cols = ['person_education', 'person_home_ownership', 'loan_intent', 'previous_loan_defaults_on_file']
data = pd.get_dummies(data, columns=categorical_cols, drop_first=True)
```

```python
data.head()
```

```python
x = data.drop("loan_status",axis=1)
y = data["loan_status"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(x_test_scaled)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(x_test_scaled)
print(classification_report(y_test, y_pred))
```
